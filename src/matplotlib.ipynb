# The `matplotlib` library

`matplotlib` is a Python library used to plot different 2-dimensional based
mathematical graphs. The `matplotlib.pyplot` sub-library has many functions to
achieve so.

### First graph

Let's try to plot our first $x$ vs. $y$ graph!


```python
import numpy as np
import matplotlib.pyplot as mp

x = np.linspace(-5, 5, 100)  # x-axis
y = x**2  # Function

print('Representing y = x**2 (a parabola)')
mp.plot(x, y)
mp.show()

x = np.linspace(-2, 2, 100)
y = np.exp(-x**2) * np.cos(2*np.pi*x)

print('Representing y = e^(-x**2) cos(2*PI*x)')
mp.plot(x, y)
mp.show()

print('Representing sinc(x) function')

np.seterr(divide='ignore', invalid='ignore')
x = np.linspace(-25, 25, 150)
y = np.sin(x) / x

mp.plot(x, y)
mp.show()
```
