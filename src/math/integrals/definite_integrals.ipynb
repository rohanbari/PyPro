# Playing with definite integrals

In simple words, integrals with its lower and upper limits are called definite
integrals. Its general form is defined as:

$$ \int*a^b {f(x)} dx = \lim*{n \to \infty} \sum*{r=1}^{n} {f(x_i^\*) (x_i - x*{i - 1})} $$


```python
# Graphical representation of an area of a function

import numpy as np
import matplotlib.pyplot as mpl


def func(x): return np.sin(0.2*x) + np.sin(2*x) + 1


x = np.linspace(0, 2*np.pi, 1000)
y = func(x)

mpl.plot(x, y)

x = np.linspace(np.pi, 1.5*np.pi, 1000)
y = func(x)

mpl.fill_between(x, y)

mpl.title('Area of a shape')
mpl.xlim([0, 2*np.pi])
mpl.ylim([0, 3])
mpl.show()
```
